```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data
file_path = 'C:/Users/PC/OneDrive/Desktop/MLBB_Draft_Data3.1.xlsx'
df = pd.read_excel(file_path, sheet_name='Sheet1')
```

```python
# Ban/Pick Analysis
ban_columns = [
    'blueTeamBan1', 'redTeamBan1', 'blueTeamBan2', 'redTeamBan2', 'blueTeamBan3', 'redTeamBan3', 
    'redTeamBan4', 'blueTeamBan4', 'redTeamBan5', 'blueTeamBan5'
]
pick_columns = [
    'blueTeamPick1', 'redTeamPick1', 'redTeamPick2', 'blueTeamPick2', 'blueTeamPick3', 'redTeamPick3', 
    'redTeamPick4', 'blueTeamPick4', 'blueTeamPick5', 'redTeamPick5'
]

ban_frequencies = df[ban_columns].apply(pd.Series.value_counts).fillna(0).sum(axis=1).sort_values(ascending=False).head(10)
pick_frequencies = df[pick_columns].apply(pd.Series.value_counts).fillna(0).sum(axis=1).sort_values(ascending=False).head(10)

# Plotting Top 10 Ban Frequencies
plt.figure(figsize=(12, 6))
sns.barplot(x=ban_frequencies.values, y=ban_frequencies.index, palette='viridis')
plt.title('Top 10 Hero Ban Frequencies', fontsize=16)
plt.xlabel('Frequency', fontsize=12)
plt.ylabel('Hero', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
for index, value in enumerate(ban_frequencies.values):
    plt.text(value, index, str(int(value)), va='center', fontsize=10)
plt.tight_layout()
plt.show()

# Plotting Top 10 Pick Frequencies
plt.figure(figsize=(12, 6))
sns.barplot(x=pick_frequencies.values, y=pick_frequencies.index, palette='viridis')
plt.title('Top 10 Hero Pick Frequencies', fontsize=16)
plt.xlabel('Frequency', fontsize=12)
plt.ylabel('Hero', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
for index, value in enumerate(pick_frequencies.values):
    plt.text(value, index, str(int(value)), va='center', fontsize=10)
plt.tight_layout()
plt.show()
```

```python
# Define the pick columns
pick_columns = {
    'blueTeamPick1': 'Blue Team 1st Pick',
    'redTeamPick1': 'Red Team 1st Pick',
    'blueTeamPick2': 'Blue Team 2nd Pick',
    'redTeamPick2': 'Red Team 2nd Pick',
    'blueTeamPick3': 'Blue Team 3rd Pick',
    'redTeamPick3': 'Red Team 3rd Pick',
    'blueTeamPick4': 'Blue Team 4th Pick',
    'redTeamPick4': 'Red Team 4th Pick',
    'blueTeamPick5': 'Blue Team 5th Pick',
    'redTeamPick5': 'Red Team 5th Pick'
}

# Plotting each pick position
for pick_col, pick_label in pick_columns.items():
    pick_frequencies = df[pick_col].value_counts().sort_values(ascending=False)
    
    plt.figure(figsize=(12, 6))
    sns.barplot(x=pick_frequencies.values, y=pick_frequencies.index, palette='viridis')
    plt.title(f'{pick_label} Frequencies', fontsize=16)
    plt.xlabel('Frequency', fontsize=12)
    plt.ylabel('Hero', fontsize=12)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    for index, value in enumerate(pick_frequencies.values):
        plt.text(value, index, str(int(value)), va='center', fontsize=10)
    plt.tight_layout()
    plt.show()
```

```python
# Define the role columns
role_columns = {
    'roleBlueTeamPick1': 'Blue Team 1st Pick Role',
    'roleRedTeamPick1': 'Red Team 1st Pick Role',
    'roleBlueTeamPick2': 'Blue Team 2nd Pick Role',
    'roleRedTeamPick2': 'Red Team 2nd Pick Role',
    'roleBlueTeamPick3': 'Blue Team 3rd Pick Role',
    'roleRedTeamPick3': 'Red Team 3rd Pick Role',
    'roleBlueTeamPick4': 'Blue Team 4th Pick Role',
    'roleRedTeamPick4': 'Red Team 4th Pick Role',
    'roleBlueTeamPick5': 'Blue Team 5th Pick Role',
    'roleRedTeamPick5': 'Red Team 5th Pick Role'
}

# Plotting each role for each pick position
for role_col, role_label in role_columns.items():
    role_frequencies = df[role_col].value_counts().sort_values(ascending=False)
    
    plt.figure(figsize=(12, 6))
    sns.barplot(x=role_frequencies.values, y=role_frequencies.index, palette='viridis')
    plt.title(f'{role_label} Frequencies', fontsize=16)
    plt.xlabel('Frequency', fontsize=12)
    plt.ylabel('Role', fontsize=12)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    for index, value in enumerate(role_frequencies.values):
        plt.text(value, index, str(int(value)), va='center', fontsize=10)
    plt.tight_layout()
    plt.show()
```

```python
# Determine wins for blue and red teams
df['BlueTeamWin'] = df.apply(lambda row: row['winningTeam'] == row['blueTeamName'], axis=1)
df['RedTeamWin'] = df.apply(lambda row: row['winningTeam'] == row['redTeamName'], axis=1)

# Count wins for blue and red teams
blue_team_wins = df['BlueTeamWin'].sum()
red_team_wins = df['RedTeamWin'].sum()

# Create a DataFrame for plotting
win_data = pd.DataFrame({
    'Team': ['Blue Team', 'Red Team'],
    'Wins': [blue_team_wins, red_team_wins]
})

# Define colors
colors = ['blue', 'red']

# Plotting the wins
plt.figure(figsize=(10, 6))
sns.barplot(x='Team', y='Wins', data=win_data, palette=colors)
plt.title('Number of Wins for Blue and Red Teams', fontsize=16)
plt.xlabel('Team', fontsize=12)
plt.ylabel('Number of Wins', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
for index, value in enumerate(win_data['Wins']):
    plt.text(index, value, str(int(value)), ha='center', va='bottom', fontsize=10)
plt.tight_layout()
plt.show()
```

```python
# Define the role columns
role_columns = {
    'roleBlueTeamPick1': 'Blue Team 1st Pick',
    'roleRedTeamPick1': 'Red Team 1st Pick',
    'roleBlueTeamPick2': 'Blue Team 2nd Pick',
    'roleRedTeamPick2': 'Red Team 2nd Pick',
    'roleBlueTeamPick3': 'Blue Team 3rd Pick',
    'roleRedTeamPick3': 'Red Team 3rd Pick',
    'roleBlueTeamPick4': 'Blue Team 4th Pick',
    'roleRedTeamPick4': 'Red Team 4th Pick',
    'roleBlueTeamPick5': 'Blue Team 5th Pick',
    'roleRedTeamPick5': 'Red Team 5th Pick'
}

# Combine role frequencies into a single DataFrame
role_data = []
for role_col, role_label in role_columns.items():
    role_counts = df[role_col].value_counts().reset_index()
    role_counts.columns = ['Role', 'Frequency']
    role_counts['Pick'] = role_label
    role_data.append(role_counts)

combined_role_data = pd.concat(role_data)

# Plotting the roles frequencies
plt.figure(figsize=(14, 8))
bar_plot = sns.barplot(data=combined_role_data, x='Pick', y='Frequency', hue='Role', palette='viridis')
plt.title('Roles Frequencies for Each Pick Position', fontsize=16)
plt.xlabel('Pick Position', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.legend(title='Role', bbox_to_anchor=(1.05, 1), loc='upper left')

# Add exact values on top of the bars
for p in bar_plot.patches:
    bar_plot.annotate(format(p.get_height(), '.0f'),
                      (p.get_x() + p.get_width() / 2., p.get_height()),
                      ha = 'center', va = 'center', 
                      xytext = (0, 9), 
                      textcoords = 'offset points',
                      fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Define the role columns and their corresponding pick positions in the correct sequence
role_columns = [
    ('roleBlueTeamPick1', 'Blue Team 1st Pick'),
    ('roleRedTeamPick1', 'Red Team 1st Pick'),
    ('roleRedTeamPick2', 'Red Team 2nd Pick'),
    ('roleBlueTeamPick2', 'Blue Team 2nd Pick'),
    ('roleBlueTeamPick3', 'Blue Team 3rd Pick'),
    ('roleRedTeamPick3', 'Red Team 3rd Pick'),
    ('roleRedTeamPick4', 'Red Team 4th Pick'),
    ('roleBlueTeamPick4', 'Blue Team 4th Pick'),
    ('roleBlueTeamPick5', 'Blue Team 5th Pick'),
    ('roleRedTeamPick5', 'Red Team 5th Pick')
]

# Combine role frequencies into a single DataFrame
role_data = []
for role_col, role_label in role_columns:
    role_counts = df[role_col].value_counts().reset_index()
    role_counts.columns = ['Role', 'Frequency']
    role_counts['Pick'] = role_label
    role_data.append(role_counts)

combined_role_data = pd.concat(role_data)

# Pivot the data to have roles as columns and picks as index
pivot_data = combined_role_data.pivot(index='Pick', columns='Role', values='Frequency').fillna(0)

# Ensure the index is in the correct order
pivot_data = pivot_data.reindex([label for _, label in role_columns])

# Define a distinct color palette
palette = sns.color_palette("hsv", len(combined_role_data['Role'].unique()))
role_colors = dict(zip(combined_role_data['Role'].unique(), palette))

# Plotting the roles frequencies using line plot
plt.figure(figsize=(14, 8))
for role in pivot_data.columns:
    plt.plot(pivot_data.index, pivot_data[role], marker='o', label=role, color=role_colors[role])

plt.title('Roles Frequencies for Each Pick Position', fontsize=16)
plt.xlabel('Pick Position', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.legend(title='Role', bbox_to_anchor=(1.05, 1), loc='upper left')

# Add exact values on top of the lines
for i, pick in enumerate(pivot_data.index):
    for role in pivot_data.columns:
        plt.text(i, pivot_data.loc[pick, role], f'{pivot_data.loc[pick, role]:.0f}', ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```
